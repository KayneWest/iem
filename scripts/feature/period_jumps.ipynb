```python
import iemdb
COOP = iemdb.connect('coop', bypass=True)
ccursor = COOP.cursor()

import numpy
import mx.DateTime

up = numpy.zeros( (10,10), 'f')
down = numpy.zeros( (10,10), 'f')
months = numpy.chararray( (10,10), itemsize='3')

ccursor.execute("""
    SELECT day, high, month from alldata_ia where day >= '1890-01-01'
    and station = 'IA2203' ORDER by day ASC
""")

revolver = []

for row in ccursor:
    revolver.insert(0, row[1])
    if len(revolver) < 23:
        continue
    
    revolver.pop()
    # up to five in front
    # i=10, j=1,2,3,4,5
    # i=10:11, j=1,2,3,4,5
    # i=10:13
    
    for i in range(11,18):
        avgi = numpy.average( revolver[10:i])
        for j in range(1,8):
            avgj = numpy.average( revolver[i:i+j])
            jump = avgi - avgj
            idx = i-10
            jdx = j
            if jump > up[idx,jdx]:
                up[idx,jdx] = jump
                #print 'Jump Up   %s i: %s j: %s %s %s %s' % (jump, idx, jdx, row[0], revolver[10:i], revolver[i:i+j])
            if jump < down[idx,jdx]:
                down[idx,jdx] = jump
                months[idx,jdx] = mx.DateTime.strptime("2000-%s-01" % (row[2],), '%Y-%m-%d').strftime("%b").upper()
                #print 'Jump Down %s i: %s j: %s %s %s %s' % (jump, idx, jdx, row[0], revolver[10:i], revolver[i:i+j])
                
```

```python
import matplotlib.pyplot as plt

(fig, ax) = plt.subplots(1,1 )

#for i in range(1,8):
#    for j in range(1,8):
#        ax.text(i, j, "%.1f" % (up[i,j],))

for j in range(1,8):
    for i in range(1,8):
        ax.text(i, j+0.2, "%.1f" % (down[i,j],), va='center', ha='center')
        ax.text(i, j-0.35, "%s" % (months[i,j], ))

for i in range(2,8):
    ax.plot([i-0.5,i-0.5], [0.5,7.5], color='k')

for j in range(2,8):
    ax.plot([0.5,7.5], [j-0.5,j-0.5], color='k')

ax.set_frame_on(False)
plt.setp(ax.get_yticklines(), visible=False)
plt.setp(ax.get_xticklines(), visible=False)
ax.set_xlim(0.5, 7.5)
ax.set_ylim(0.5, 7.5)
ax.set_ylabel("Period of Days Prior")
ax.set_xlabel("Period of Inclusive Days After")
ax.set_title("1880-2012 Des Moines Max of Avg High Temp Drop $^{\circ}\mathrm{F}$\nfrom a period of days to another period of days")

fig.savefig('test.ps')
import iemplot
iemplot.makefeature('test')
```
