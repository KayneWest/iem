```python
import psycopg2
pgconn = psycopg2.connect(database='postgis', host='iemdb', user='nobody')
cursor = pgconn.cursor()

cursor.execute("""
  SELECT ugc, max(issue) from warnings where phenomena = 'TO' and significance = 'W'
  and substr(ugc,1,2) = 'IA' GROUP by ugc ORDER by max ASC""")

data = {}
labels = {}
for row in cursor:
    print row
    labels[ row[0] ] = row[1].strftime("%-d %b")
    data[ row[0] ] = int(row[1].year)
```

```python
from pyiem.plot import MapPlot
import matplotlib.cm as cm

cmap = cm.get_cmap('jet')
cmap.set_over('k')
cmap.set_under('#EEEEEE')

m = MapPlot(title='Year of Last Tornado Warning by Iowa County',
            subtitle='Through 5 AM 12 May 2014')
m.fill_ugc_counties(data, range(2008,2016), cmap=cmap, units='year')
m.drawcounties()
m.postprocess(filename='test.ps')
m.close()
```
