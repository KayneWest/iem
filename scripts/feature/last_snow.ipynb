```python
import iemdb
COOP = iemdb.connect('coop', bypass=True)
cursor = COOP.cursor()
cursor2 = COOP.cursor()

lastdoys = []
lastyears = []
tempyears = []
maxtemps = []
cursor.execute("""SELECT max(extract(doy from day)), max(day), year from alldata where
  station = 'IA2203' and snow >= 1 and month < 7 GROUP by year ORDER by year ASC""")
for row in cursor:
    lastdoys.append( row[0] )
    lastyears.append( row[2] )
    cursor2.execute("""SELECT max(high) from alldata where station = 'IA2203' and year = 
     %s and day < %s """, (row[2], row[1]) )
    row2 = cursor2.fetchone()
    tempyears.append( row[2] )
    maxtemps.append( row2[0] )
```

```python
import matplotlib.pyplot as plt
import numpy
lastdoys = numpy.array( lastdoys )
maxtemps = numpy.array(maxtemps )

(fig, ax) = plt.subplots(2,1, sharex=True)

ax[0].scatter(lastyears, lastdoys)
ax[0].plot([1880,2014], [numpy.average(lastdoys), numpy.average(lastdoys)] , c='r', lw=2.0)
ax[0].set_yticks( (1,32,60,91,121,152,182,213,244,274,305,335,365) )
ax[0].set_yticklabels( ('Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec') )
ax[0].grid(True)
ax[0].set_xlim(1880, 2014)
ax[0].set_ylim(0,130)
ax[0].set_title("Des Moines Date of Last 1+ inch Snowfall")

ax[1].plot([1880,2014], [numpy.average(maxtemps), numpy.average(maxtemps)] , c='r', lw=2.0)
ax[1].bar(tempyears, maxtemps)
ax[1].set_ylim(40,90)
ax[1].grid(True)
ax[1].set_ylabel("YTD Max Temperature $^{\circ}\mathrm{F}$")
ax[1].set_xlabel("*2013 High to date is 86, ploted max is: %s" % (numpy.max(maxtemps),))

fig.savefig('test.ps')
import iemplot
iemplot.makefeature('test')
```
