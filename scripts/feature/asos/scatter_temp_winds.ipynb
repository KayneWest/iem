```python
import psycopg2
pgconn = psycopg2.connect(database='asos', host='iemdb', user='nobody')
cursor = pgconn.cursor()

cursor.execute("""
 WITH climo as (
   SELECT extract(day from valid) as d, avg(tmpf) from alldata
   WHERE station = 'DSM' and tmpf is not null and extract(month from valid) = 4
   and extract(hour from valid + '10 minutes'::interval) = 15 GROUP by d
 ), obs as (
   SELECT valid, extract(day from valid) as d, tmpf, sknt from alldata WHERE station = 'DSM' and extract(month from valid) = 4
   and extract(hour from valid + '10 minutes'::interval) = 15 and tmpf is not null and
   sknt is not null
 )

 SELECT o.valid, o.tmpf - c.avg, o.sknt from obs o JOIN climo c on (c.d  = o.d) ORDER by o.valid ASC

""")
sknt = []
deltat = []
for row in cursor:
    deltat.append( row[1] )
    sknt.append( row[2] * 1.15)
```

```python
import numpy as np
sknt = np.array(sknt)
deltat = np.array(deltat)

H, xedges, yedges = np.histogram2d(deltat, sknt, bins=[30,20], normed=True, range=[[-30,30], [0,40]])
extent = [xedges[0], xedges[-1], yedges[-1], yedges[0]]

avgy = []
for i in range(30):
    avgy.append( np.sum( H[i,:] * yedges[:-1] ) / np.sum(H[i,:]) )
avgy = np.array(avgy) 
```

```python
print deltat[-1], sknt[-1]
import matplotlib.pyplot as plt
import matplotlib.cm as cm

(fig, ax) = plt.subplots(1,1)

#ax.scatter(deltat, sknt * 1.15, facecolor='b', edgecolor='b')
#print extent
res = ax.imshow(H.transpose(),  extent=extent, aspect='auto', interpolation='nearest',
                cmap=cm.get_cmap('bone_r'), label='Frequency')
ax.plot( np.linspace(-30,30,30), avgy, color='red', lw=3.5, zorder=2, label='Average')
ax.plot( np.linspace(-30,30,30), avgy, color='white', lw=2, zorder=3)
ax.scatter(deltat[-1], sknt[-1], marker='o', s=90, color='brown', label='9 Apr 2014')
ax.set_ylim(0,40)
ax.grid(True)
ax.set_xlim(-30,30)
ax.legend(ncol=2)
ax.set_ylabel("Wind Speed [mph]")
ax.set_xlabel("Temperature Departure $^\circ$F, relative frequency shaded")
ax.set_title("1936-2013 Des Moines April at 3 PM Composite\nWind Speed v. Temperature Departure from Average")
#clr = fig.colorbar(res)
```
