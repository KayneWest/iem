```python
import psycopg2
from pyIEM import mesonet
import math
import numpy as np

ASOS = psycopg2.connect(database="asos", host='mesonet.agron.iastate.edu', user='nobody')
acursor = ASOS.cursor()
def get_data(tbl):
    acursor.execute("""SELECT extract(doy from valid) as d, dwpf from """+tbl+"""
        where station = 'DSM' and dwpf > -50 and dwpf < 90 and 
        tmpf > -50 and tmpf < 120 and valid > '1950-01-01' """)

    dtot = numpy.zeros( (366,), 'f')
    dmin = numpy.zeros( (366,), 'f')
    dmax = numpy.zeros( (366,), 'f')
    dcnt = numpy.zeros( (366,), 'f')
    dmin[:] = 100.

    for row in acursor:
        dwpc = mesonet.f2c( row[1] )
        e  = 6.112 * math.exp( (17.67 * dwpc) / (dwpc + 243.5));
        mixr =  0.62197 * e / (1000.0 - e)
        if mixr > dmax[row[0] -1]:
            dmax[row[0] -1] = mixr
        if mixr < dmin[row[0] -1]:
            dmin[row[0] -1] = mixr

        dtot[ row[0] -1] += mixr
        dcnt[ row[0] -1] += 1

    return dtot, dmin, dmax, dcnt
        
dtot, dmin, dmax, dcnt = get_data('alldata')
dtot13, dmin13, dmax13, dcnt13 = get_data('t2013')
```

```python
import matplotlib.pyplot as plt

climate = np.array(dtot) / np.array(dcnt)
d2013 = np.array(dtot13) / np.array(dcnt13)
dmax = numpy.array(dmax)
dmin = numpy.array(dmin)

fig = plt.figure()
ax = fig.add_subplot(211)

ax.fill_between(np.arange(0,366), dmin * 1000., dmax * 1000., color='gray')

ax.plot(np.arange(0,366), climate * 1000., label="Climatology")
ax.plot(np.arange(0,len(d2013)), d2013 * 1000., color='r', label="2013")

ax.set_title("Des Moines [KDSM] Daily Mean Surface Water Vapor Mixing Ratio")
ax.set_ylabel("Mixing Ratio ($g/kg$)")
ax.set_xlim(0,366)
ax.set_ylim(0, 26.)
ax.set_xticks( (1,32,60,91,121,152,182,213,244,274,305,335,365) )
ax.set_xticklabels( ('Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec') )
ax.grid(True)
ax.legend(loc=2, fontsize=10)

ax = fig.add_subplot(212)

rects = ax.bar(numpy.arange(0,len(d2013)), (d2013  - climate[:len(d2013)]) * 1000.0, edgecolor='b')
for rect in rects:
  if rect.get_y() < 0.:
    rect.set_facecolor('r')
    rect.set_edgecolor('r')
  else:
    rect.set_facecolor('b')

ax.set_ylabel("2013 Departure ($g/kg$)")
ax.set_xlim(0,366)
ax.set_xticks( (1,32,60,91,121,152,182,213,244,274,305,335,365) )
ax.set_xticklabels( ('Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec') )
ax.grid(True)

import iemplot
fig.savefig('test.ps')
iemplot.makefeature('test')
```
