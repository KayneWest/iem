```python
import psycopg2
import pandas as pd
from pyiem import network
import matplotlib.pyplot as plt

nt = network.Table(('IA_ASOS',))
```

```python
COOPDB = psycopg2.connect(database='coop', host='iemdb', user='nobody')
IEMDB = psycopg2.connect(database='iem', host='iemdb', user='nobody')
cursor = IEMDB.cursor()
ccursor = COOPDB.cursor()

cursor.execute("""SELECT id, avg((max_tmpf+min_tmpf)/2.)
  from summary_2014 s JOIN stations t on (t.iemid = s.iemid) WHERE
  t.network in ('IA_ASOS') and day >= '2014-07-01' and day < '2014-07-28' 
  and max_tmpf < 120 GROUP by id ORDER by avg ASC""")

res = []
for row in cursor:
    if not nt.sts.has_key(row[0]):
        continue
    climatesite = nt.sts[row[0]]['climate_site']
    ccursor.execute("""SELECT sum(case when avg < %s then 1 else 0 end), count(*) from (
    SELECT year, avg((high+low)/2.) from 
    alldata_"""+climatesite[:2]+""" where station = %s and year < 2014
    and sday < '0728' and month = 7
    GROUP by year) as foo""", (row[1], climatesite) )
    row2 = ccursor.fetchone()
    
    res.append( dict(label=row[0], val=row[1], rank=(row2[0]+1), years=row2[1]) )
    #print "%s %.1f %s %s" % (row[0], row[1], row2[0], row2[1])
    #lons.append( nt.sts[row[0]]['lon'] )
    #lats.append( nt.sts[row[0]]['lat'] )
    #vals.append( row[1] - row2[0] )

df = pd.DataFrame(res)
    
```
