```python
import numpy as np
import datetime
import psycopg2
ASOS = psycopg2.connect(database='asos', host='iemdb', user='nobody')
cursor = ASOS.cursor()

IEM = psycopg2.connect(database='iem', host='iemdb', user='nobody')
icursor = IEM.cursor()

precip = {}
icursor.execute("""SELECT date(valid) as d, sum(phour) from hourly
  WHERE station = 'DSM' and extract(month from valid) in (7,8) AND
  phour > 0 and extract(hour from valid) < 8 GROUP by d""")
for row in icursor:
    precip[ row[0].strftime("%Y%m%d") ] = row[1]


clrcounts = np.zeros( (24), 'f')
clrhits = np.zeros( (24), 'f')

cursor.execute("""
SELECT distinct to_char(valid, 'YYYYmmddHH24') from alldata where
station = 'DSM' and 'CLR' in (skyc1, skyc2, skyc3, skyc4) and
  extract(month from valid) = 7 
""")
for row in cursor:
    ts = datetime.datetime.strptime(row[0], '%Y%m%d%H')
    clrcounts[ts.hour - 1] += 1
    if precip.get((ts + datetime.timedelta(days=1)).strftime("%Y%m%d"), 0) > 0:
        clrhits[ts.hour ] += 1
        

```
