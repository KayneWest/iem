```python
import psycopg2
import numpy
ASOS = psycopg2.connect(database='asos', user='nobody', host='iemdb')
cursor = ASOS.cursor()

tmp_hits = numpy.zeros( (121-50,), 'f')
tmp_cnts = numpy.zeros( (121-50,), 'f')
doy_hits = numpy.zeros( (366,), 'f')
doy_cnts = numpy.zeros( (366,), 'f')

cursor.execute("""SELECT extract(doy from valid), round(tmpf::numeric,0), dwpf from alldata
  WHERE station = 'DSM' and extract(hour from valid + '10 minutes'::interval) in (13,14,15,16)
  and tmpf is not null and dwpf is not null""")
for row in cursor:
    doy_cnts[ row[0] - 1 ] += 1.0
    if row[1] >= 55:
        tmp_cnts[ row[1] - 55 ] += 1.0
        if row[2] < 55:
            tmp_hits[ row[1] - 55 ] += 1.0
    if row[2] < 55:
        doy_hits[ row[0] - 1 ] += 1.0
        
cursor.close()
ASOS.close()
```

```python
import matplotlib.pyplot as plt
(fig, ax) = plt.subplots(2,1)
doy_cnts = numpy.where( doy_cnts < 1, 1, doy_cnts)

ax[0].bar( numpy.arange(55,101)-0.4, tmp_hits[:46] / tmp_cnts[:46] * 100.0, width=1.0,
  fc='r', ec='r')
ax[0].set_xlim(54.5,100.5)
ax[0].grid(True)
ax[0].set_title("1933-2012 Des Moines Afternoon (1-4 PM) Frequency\nof sub 55$^{\circ}\mathrm{F}$ Dew Points")
ax[0].set_ylabel("Freq.[%] by Air Temp")
ax[0].set_yticks([0,25,50,75,100])

ax[1].bar( numpy.arange(50,321), doy_hits[50:321] / doy_cnts[50:321] * 100.0, width=1.0,
  fc='g', ec='g')
ax[1].set_xticks( (1,32,60,91,121,152,182,213,244,274,305,335,365) )
ax[1].set_xticklabels( ('Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec') )
ax[1].set_xlim(50,320)
ax[1].grid(True)
ax[1].set_ylabel("Freq.[%] by Day of Year")
ax[1].set_yticks([0,25,50,75,100])

#fig.savefig('test.ps')
#import iemplot
#iemplot.makefeature('test')
```
