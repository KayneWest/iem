```python
import psycopg2
ASOS = psycopg2.connect(database='asos', host='iemdb', user='nobody')
cursor = ASOS.cursor()

cursor.execute("""
SELECT sixam.tmpf, twopm.tmpf, sixam.week, sixam.date from 
 (SELECT extract(week from valid) as week, date(valid), tmpf from alldata WHERE
  station = 'DSM' and extract(hour from valid + '10 minutes'::interval) = 6
  and extract(minute from valid + '10 minutes'::interval) < 11
  and tmpf is not null) as sixam,
 (SELECT extract(week from valid) as week, date(valid), tmpf from alldata WHERE
  station = 'DSM' and extract(hour from valid + '10 minutes'::interval) = 14
  and extract(minute from valid + '10 minutes'::interval) < 11
  and tmpf is not null) as twopm
  WHERE twopm.date = sixam.date ORDER by (sixam.tmpf - twopm.tmpf) DESC
""")

weeks = []
difference = []
for row in cursor:
    weeks.append( row[2] )
    difference.append( float(row[1] - row[0]) )
    #print "date: %s 6 AM: %.1f 2 PM: %.1f" % (row[3], row[0], row[1])
```
