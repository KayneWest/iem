```python
%pylab inline
import psycopg2
```

```python
IEM = psycopg2.connect(database='iem', host='iemdb', user='nobody')
cursor = IEM.cursor()

times = []
vals = []
cursor.execute("""SELECT valid, tmpf from current_log c JOIN stations t ON (t.iemid = c.iemid)
 WHERE t.network in ('IA_ASOS', 'AWOS') and tmpf is not null""")
for row in cursor:
    times.append(row[0])
    vals.append( row[1] )
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pytz

(fig, ax) = plt.subplots(1,1, figsize=(8,7))
ax.scatter(times, vals, edgecolor='None', alpha=0.2)
ax.set_title("Iowa Airport Temperature Reports\nASOS/AWOS stations: 12 AM 21 Mar - 5 AM 23 Mar 2013")
ax.set_ylabel("MSLP [mb]")
ax.grid(True)
ax.xaxis.set_major_formatter(mdates.DateFormatter('%I %p\n%-d %b', 
                                                  tz=pytz.timezone("America/Chicago")))
ax.set_xlim(min(times), max(times))
#ax.set_ylim(bottom=29, top=30.5)
fig.savefig('150323.png')
```
