```python
import psycopg2
POSTGIS = psycopg2.connect(database='postgis', host='iemdb', user='nobody')
pcursor = POSTGIS.cursor()

events = []
pcursor.execute("""SELECT issue, expire, significance from warnings
 WHERE ugc = (select ugc from ugcs where name = 'Black Hawk' and state = 'IA' LIMIT 1)
 and phenomena = 'WC' and significance != 'A' and issue > '2013-12-20' ORDER by issue ASC""")
for row in pcursor:
    print row[0], row[1], row[2]
    events.append( row )

IEM = psycopg2.connect(database='asos', host='iemdb', user='nobody')
icursor = IEM.cursor()

icursor.execute("""SELECT valid, wcht(tmpf, sknt) from alldata where station = 'ALO'
 and sknt >= 0 and valid > '2013-12-20' ORDER by valid ASC""")
valid = []
wcht = []
for row in icursor:
    wcht.append( row[1] )
    valid.append( row[0] )
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Rectangle
import matplotlib.dates as mdates
import pytz

(fig, ax) = plt.subplots(1,1, figsize=(10,7))

colors = {'W': '#ff0000', 'Y': '#00ff00'}

ax.plot(valid, wcht, zorder=2, lw=2)
ymin, ymax = ax.get_ylim()
for event in events:
    delta = (event[1] - event[0]).days * 86400.0 + (event[1] - event[0]).seconds
    rect = Rectangle( (event[0], ymin), delta / 86400.0, ymax-ymin, alpha=0.5, facecolor=colors[event[2]], edgecolor='none', zorder=1)
    ax.add_patch(rect)
    
ax.set_xlim(min(valid), events[-1][1])
ax.xaxis.set_major_locator(
                               mdates.DayLocator(interval=7,
                                                 tz=pytz.timezone("America/Chicago")))
                               
ax.xaxis.set_major_formatter(mdates.DateFormatter('%-d\n%b',
                                                      tz=pytz.timezone("America/Chicago")))
ax.grid(True)
ax.set_ylabel("Wind Chill Temperature $^\circ$F", fontsize=16)
ax.set_title("20 Dec 2013 - 6 Feb 2014 Waterloo Wind Chill\nNWS Issued Wind Chill Advisories(green) + Warnings(red)", fontsize=16)
fig.savefig('140206.png')
```
