```python
%pylab inline
import numpy as np

import psycopg2
ASOS = psycopg2.connect(database='asos', host='iemdb', user='nobody')
cursor = ASOS.cursor()

COOP = psycopg2.connect(database='coop', host='iemdb', user='nobody')
ccursor = COOP.cursor()

climate = {}
ccursor.execute("""SELECT valid, high from climate51 where station= 'IA2203'""")
for row in ccursor:
    climate[ row[0].strftime("%m%d") ] = row[1]


counts = np.zeros( (53), 'f')
hits = np.zeros( (53), 'f')

cursor.execute("""
WITH FOG as
 (SELECT distinct date(valid) from alldata where station = 'DSM' and 
 extract(hour from valid + '10 minutes'::interval) = 6 and presentwx ~* 'FG'
 and valid > '1971-01-01'),
HIGHS as 
(SELECt date(valid), max(tmpf) from alldata where station = 'DSM' and
 extract(hour from valid) between 12 and 19 and tmpf > -50 GROUP by date)
 SELECT FOG.date, HIGHS.max from FOG JOIN HIGHS on (FOG.date = HIGHS.date)
""")
for row in cursor:
    week = int(row[0].isocalendar()[1])-1
    if row[1] > climate[ row[0].strftime("%m%d") ]:
        hits[week] += 1
    counts[week] += 1

ASOS.close()
```

```python
import matplotlib.pyplot as plt

(fig, ax) = plt.subplots(1,1)

ax.bar(np.arange(1,54)-0.4, hits / counts * 100.0 )
ax.set_xticks( np.arange(1,55,7) )
ax.set_xticklabels( ('Jan 1', 'Feb 19', 'Apr 8', 'May 27', 'Jul 15', 'Sep 2', 'Oct 21', 'Dec 9'))
ax.grid(True)
ax.set_xlim(0.5,52.5)
ax.set_title("1971-2013 Des Moines when 6 AM fog was reported,\nwas the afternoon high above average?")
ax.set_ylabel("Percent of Days [%]")
ax.set_xlabel("Partitioned by week of year")
```
