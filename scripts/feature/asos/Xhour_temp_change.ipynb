```python
%pylab inline
import psycopg2
import network
ASOS = psycopg2.connect(database='asos', user='nobody',host='iemdb')
acursor = ASOS.cursor()

#nt = network.Table(('IA_ASOS','MO_ASOS','IL_ASOS', 'ND_ASOS', 'AWOS',
#          'WI_ASOS','MN_ASOS', 'SD_ASOS', 'NE_ASOS', 'KS_ASOS',
#          'IN_ASOS','KY_ASOS','OH_ASOS','MI_ASOS'))
nt = network.Table(('IA_ASOS', 'AWOS'))
lats = []
lons = []
vals = []
for sid in nt.sts.keys():
    if nt.sts[sid]['archive_begin'] is None or nt.sts[sid]['archive_begin'].year > 1973:
        continue

    acursor.execute("""
    with 
    one as (select tmpf, valid from t2013 where station = %s 
    and tmpf > -40), 
    two as (select tmpf, valid - '24 hours'::interval as valid2, valid from t2013 
    where station = %s and tmpf > -40) 
    
    select one.valid, two.valid, one.tmpf, two.tmpf, one.tmpf - two.tmpf as diff 
    from one JOIN two on (one.valid = two.valid2) WHERe one.tmpf > two.tmpf + 30 
    ORDER by diff DESC LIMIT 10
    """, (sid, sid))
    
    myvals = []
    for row in acursor:
        myvals.append( row[4] )
        if len(myvals) < 3:
            print sid, row[0].strftime("%m/%d/%Y %H:%M"), row[1].strftime("%m/%d/%Y %H:%M"), row[2], row[3], row[4]
    if len(myvals) < 10:
        continue
    if myvals[0] - myvals[1] > 20:
        #print 'BAD', sid, myvals
        myvals[0] = myvals[1]

    if myvals[0] < 150:
        lats.append( nt.sts[sid]['lat'] )
        lons.append( nt.sts[sid]['lon'] )
        vals.append( myvals[0] )
        #print sid, myvals[:3]
```

```python
from pyiem.plot import MapPlot

lats2 = []
lons2 = []
vals2 = []

for (v,la,lo) in zip(vals, lats, lons):
    if v > 80:
        print v, la, lo
        continue
    lats2.append( la )
    lons2.append( lo )
    vals2.append( v )

m = MapPlot(sector='midwest', title="Maximum 24 Hour Air Temperature Change $^\circ$F",
            subtitle="based on automated stations dating back to at least 1971")
m.plot_values(lons2, lats2, vals2, fmt='%.0f')
m.postprocess(filename='test.png')

```
