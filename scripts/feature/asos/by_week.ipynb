```python
import psycopg2
import matplotlib.pyplot as plt
import numpy as np
```

```python
PGCONN = psycopg2.connect(database='asos', host='iemdb', user='nobody')
cursor = PGCONN.cursor()

def get(station):
        cursor.execute("""select extract(week from valid) as w, count(*),
        sum(case when dwpf < 50 then 1 else 0 end) / count(*)::numeric from alldata 
        where station = %s and tmpf >= 79.5 GROUP by w ORDER by w ASC""", (station,))
        weeks = []
        freq = []
        for row in cursor:
            if row[1] < 10:
                continue
            weeks.append( row[0] )
            freq.append( row[2] )
            
        return np.array(weeks), np.array(freq)
    
#mcw_weeks, mcw_freq = get('MCW')
mob_weeks, mob_freq = get('MOB')
 
```
