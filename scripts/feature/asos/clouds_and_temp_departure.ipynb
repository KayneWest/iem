```python
import iemdb
import numpy
import numpy.ma
ASOS = iemdb.connect('asos', bypass=True)
acursor = ASOS.cursor()

tclimo = numpy.zeros( (24, 52), 'f')
counts= numpy.zeros( (24, 52), 'f')
diff = numpy.ma.zeros( (24, 52), 'f')
```

```python
acursor.execute("""
 select extract(week from valid) as w, 
 extract(hour from valid + '10 minutes'::interval) as hr, 
 avg(tmpf), min(tmpf) from alldata where station = 'DSM' and tmpf is not null 
 and tmpf > -99 GROUP by w, hr
""")
for row in acursor:
    if row[0] == 53:
        continue
    tclimo[ row[1], row[0]-1 ] = row[2]
    counts[ row[1], row[0]-1 ] = row[3]
```

```python
acursor.execute("""
 select extract(week from valid) as w, 
 extract(hour from valid + '10 minutes'::interval) as hr,
 avg(tmpf), count(*) from alldata where station = 'DSM' and tmpf is not null 
 and (skyc1 in ('BKN','OVC') or skyc2 in ('BKN','OVC') or 
        skyc3 in ('BKN','OVC')) and tmpf > -99 
 GROUP by w, hr
""")
for row in acursor:
    if row[0] == 53:
        continue
    d = row[2] - tclimo[ row[1], row[0]-1 ]
    if d > 10:
        print row, tclimo[ row[1], row[0]-1 ], counts[ row[1], row[0]-1 ]
    diff[ row[1], row[0]-1 ] = d
```

```python
diff.mask = numpy.where(diff == 0, True, False)
import matplotlib.pyplot as plt

(fig, ax) = plt.subplots(1,1, figsize=(10,8))

cs = ax.imshow(diff, aspect='auto', interpolation='nearest')
a = fig.colorbar(cs)
a.ax.set_ylabel('Temperature Departure $^{\circ}\mathrm{F}$', fontsize=18)
ax.grid(True)
ax.set_title("Des Moines 1933-2012 Hourly Temperature Departure\nfor hours when skies were mostly cloudy, vs all", fontsize=18)
ax.set_ylim(-0.5,23.5)
ax.set_ylabel("Local Hour of Day, either CDT/CST", fontsize=18)
ax.set_yticks((0,4,8,12,16,20))
ax.set_xticks( numpy.arange(0,55,7))
ax.set_xticklabels( ('Jan 1', 'Feb 19', 'Apr 8', 'May 27', 'Jul 15', 'Sep 2', 'Oct 21', 'Dec 9'), fontsize=18,
rotation=90)

ax.set_yticklabels( ('Mid', '4 AM', '8 AM', 'Noon', '4 PM', '8 PM'),
    fontsize=18)

#fig.savefig("/tmp/121108.png")
```
