```python
import iemdb
from pyiem.nws import vtec
POSTGIS = iemdb.connect('postgis', bypass=True)
pcursor = POSTGIS.cursor()

pcursor.execute("""select phenomena, significance, min(issue), count(*) from warnings where gtype = 'C' and issue > '2009-01-01' 
 and issue < '2013-01-01' GROUP by phenomena, significance ORDER by count DESC""")

labels = []
vals = []
cnt = 1
for row in pcursor:
    l = "%s. %s %s (%s.%s)" % (cnt, 
        vtec._phenDict.get(row[0], row[0]), vtec._sigDict[row[1]], row[0], row[1])
    if cnt < 26 or row[0] in ['ZR', 'IS']:
        labels.append(l)
        vals.append( row[3] )
    cnt += 1
```

```python
import matplotlib.pyplot as plt
import numpy
(fig, ax) = plt.subplots(1,1, figsize=(7,10))
vals = numpy.array(vals)

ax.barh(numpy.arange(len(vals))-0.4, vals / float(vals[0]) * 100.0)
for i in range(1, len(vals)):
    y = vals[i] / float(vals[0]) * 100.0
    ax.text(y + 1, i, '%.1f%%' % (y,), va='center')
ax.set_title("2009-2012 NWS Watch/Warning/Advisory Totals")
ax.set_xlabel("Event+County/Zone Count, Relative to Severe Thunderstorm Warning")
ax.set_ylim(len(vals), -0.5)
ax.grid(True)
ax.set_yticklabels( labels )
ax.set_yticks( numpy.arange(len(vals)))
fig.savefig('test.png')
#import iemplot
#iemplot.makefeature('test')
```
