```python
import iemdb
COOP = iemdb.connect('coop', bypass=True)
import mx.DateTime
import numpy
```

```python
ccursor = COOP.cursor()
obs = numpy.zeros((2013-1900,366), 'f')
climate = numpy.zeros((366), 'f')
ccursor.execute("""SELECT extract(doy from valid), precip from climate
 WHERE station = 'IA1319'""")
for row in ccursor:
    climate[row[0]-1] = row[1]

ccursor.execute("""SELECT day, precip from alldata_ia where
 station = 'IA1319' and precip > 0 and year > 1899""")
for row in ccursor:
    obs[row[0].year-1900,int(row[0].strftime("%j"))-1] = row[1]
```

```python
climate7 = numpy.zeros((366),'f')
for i in range(6,366):
    climate7[i] = numpy.sum(climate[i-6:i+1])
```

```python
summers = []
for yr in range(1900,2013):
    cnt = 0
    for dy in range(121,227):
        if numpy.sum(obs[yr-1900,dy-6:dy+1]) > climate7[dy]:
            cnt += 1
            if yr == 2012:
                print dy, climate7[dy], obs[yr-1900,dy-6:dy+1]
    summers.append(cnt)
```
