```python
import psycopg2
import numpy

obs = numpy.zeros( (2014-1893, 366), 'f')

COOP = psycopg2.connect(database='coop', host='iemdb', user='nobody')
cursor = COOP.cursor()

cursor.execute("""SELECT year, extract(doy from day), gdd50(high, low) from alldata_ia
  WHERE station = 'IA0200' and year > 1892""")
for row in cursor:
    obs[ row[0] - 1893, row[1] - 1 ] = row[2]
```

```python
import matplotlib.pyplot as plt
avg = numpy.average(obs[:-1,:], 0)
(fig, ax) = plt.subplots(1,1)
ax.plot(numpy.arange(1,367), avg, color='r', zorder=2, lw=2.,
label='Average')
ax.scatter(numpy.arange(1,142), obs[-1,:141], color='b',  zorder=2,
label='2013')
p5, p25,p75,p95 = numpy.percentile(obs[:-1,:], [5,25,75,95], 0)
#ax.plot(numpy.arange(1,367), p5, color='g', zorder=2)
#ax.plot(numpy.arange(1,367), p95, color='b', zorder=2)
ax.bar(numpy.arange(1,367), p95-p5, bottom=p5, ec='tan', fc='tan', zorder=1,
label='5-95 Percentile')
ax.bar(numpy.arange(1,367), p75-p25, bottom=p25, ec='lightblue', fc='lightblue', zorder=1,
label='25-75 Percentile')
ax.set_xlim(1,367)
ax.set_ylim(-0.25,40)
ax.grid(True)
ax.set_title("1893-2013 Ames Daily Growing Degree Days")
ax.set_ylabel("Daily Accumulation $^{\circ}\mathrm{F}$")
ax.set_xticks( (1,32,60,91,121,152,182,213,244,274,305,335,365) )
ax.legend(ncol=2)
ax.set_xticklabels( ('Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec') )
fig.savefig('test.svg')
import iemplot
iemplot.makefeature('test')
```
