```python
import iemdb
import numpy
COOP = iemdb.connect('coop', bypass=True)
ccursor = COOP.cursor()
```

```python
import datetime
def year(yr):
    running = [0]
    ccursor.execute("""SELECT o.day, o.high - c.high from alldata_ia o, climate51 c where
 o.station = 'IA2203' and o.station = c.station 
 and to_char(c.valid, 'mmdd') = o.sday and year = %s ORDER by day""" % (yr,) )
    for row in ccursor:
        if row[1] > 0 and running[-1] < 0:
            running.append(1)
        elif row[1] > 0 and running[-1] > 0:
            running.append( running[-1] + 1)
        elif row[1] < 0 and running[-1] > 0:
            running.append(-1)
        else:
            running.append( running[-1] - 1)
    return running

run2012 = year(2012)
run1988 = year(1988)
run1936 = year(1936)
run1934 = year(1934)
```
