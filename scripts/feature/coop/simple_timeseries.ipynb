```python
import psycopg2
COOP = psycopg2.connect(database='coop', host='iemdb', user='nobody')
cursor = COOP.cursor()

cursor.execute("""
 WITH running as (select day, sday, avg((high+low)/2.) OVER (ORDER by day ASC ROWS between 91 preceding and current row) from alldata_ia where station = 'IA0200')  SELECT day, avg, mavg from (SELECT avg, day, avg(avg) OVER (PARTITION by sday) as mavg from running r where day >= '1893-01-01') as foo WHERE day > '2013-09-01' ORDER by day ASC
""")

days = []
mavg = []
avg = []
for row in cursor:
    days.append( row[0] )
    mavg.append( row[2] )
    avg.append( row[1] )
```

```python
import matplotlib.pyplot as plt
import datetime
from matplotlib.patches import Rectangle

(fig, ax) = plt.subplots(1,1)

ax.plot(days, mavg, zorder=2, lw=2, label="Climatology")
rect = plt.Rectangle((days[-97],0),91, 80, fc='#EEEEEE', ec='#EEEEEE', zorder=1)
ax.plot(days, avg, zorder=2, lw=2, label="2013-2014")
ax.grid(True)
ax.set_title("Ames Trailing 91 Day Average Temperature")
ax.set_ylabel("Average Temperature $^\circ$F")
ax.text(days[-51], 55, 'Coldest 91 Day Period', color='g', ha='center', fontsize=16)
ax.set_xlim(left=days[30])
ax.set_xlabel(" * 2014 thru 11 March")
ax.legend(ncol=2)
ax.add_patch(rect)
```
