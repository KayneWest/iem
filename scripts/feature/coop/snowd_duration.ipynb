```python
import numpy as np
import psycopg2
COOP = psycopg2.connect(database='coop', user='nobody', host='iemdb')
cursor = COOP.cursor()
cursor2 = COOP.cursor()

cursor.execute("""
 SELECT day from alldata_ia where station = 'IA2203' and sday = '1209' and snowd > 0
""")

counts = np.zeros( (120), 'f')

for row in cursor:
    cursor2.execute("""SELECT min(day) from alldata_ia where station = 'IA2203' and 
    snowd = 0 and day > %s""", (row[0],))
    row2 = cursor2.fetchone()
    days = (row2[0] - row[0]).days
    counts[:days] += 1
```

```python
import datetime
sts = datetime.datetime(2013,12,9)
x = []
for i in range(120):
    x.append( sts + datetime.timedelta(days=i) )
    
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

(fig, ax) = plt.subplots(1,1)

ax.xaxis.set_major_locator(
                               mdates.MonthLocator(interval=1)
                               )
ax.xaxis.set_major_formatter(mdates.DateFormatter('%-d %b'))
ax.set_yticks([0,25,50,75,100])
ax.grid(True)
ax.set_title("1893-2012 Des Moines Snowcover Duration\n With snowcover on 9 December (45/120 years)")
ax.set_ylabel("Percent of Years [%]")
ax.axvline(x[16], lw=2, c='r')
ax.text(x[15], 10, 'Dec 25', ha='right', color='r')
ax.set_xlabel("Date when 9 December snowfall disappeared")

ax.plot(x, counts / 45.0 * 100)
```
