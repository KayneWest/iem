```python
import iemdb
import numpy
COOP = iemdb.connect('coop', bypass=True)
ccursor = COOP.cursor()
ccursor2 = COOP.cursor()
```

```python
import datetime
ccursor.execute("""SELECT distinct station from alldata_ia where
 year = 1934 and precip > 0""")
counts = numpy.zeros( (2013-1893, 123), 'f')
for row in ccursor:
    station = row[0]
    precip = numpy.zeros((2013-1893, 123), 'f')
    ccursor2.execute("""SELECT year, day, precip from alldata_ia
    WHERE precip > 0 and station = '%s' and month in (5,6,7,8)""" % (station,))
    for row2 in ccursor2:
        yindex = int(row2[0] - 1893)
        xindex = (row2[1] - datetime.date(row2[0], 5, 1)).days
        precip[yindex,xindex] = row2[2]
    for day in range(30,124):
        ytd = numpy.sum( precip[:,day-30:day], 1)
        counts[:,day-1] += numpy.where( ytd < 2, 1, 0)
```
