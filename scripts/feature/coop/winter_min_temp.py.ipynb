```python
import psycopg2
COOP = psycopg2.connect(database='coop', host='iemdb', user='nobody')
cursor = COOP.cursor()

hdoy = []
hval = []
ldoy = []
lval = []
cursor.execute("""
  WITH ranks as (
  SELECT extract(year from day + '4 months'::interval) as yr, extract(doy from day) as doy, low, 
  rank() OVER (PARTITION by extract(year from day + '4 months'::interval) ORDER by low ASC) from alldata_ia where station = 'IA2203'
  and year > 1892
  ORDER by yr ASC
  )

  SELECT yr, doy, low from ranks WHERE rank = 1

""")
for row in cursor:
    addon = 366 if row[1] < 180 else 0
    ldoy.append( row[1] + addon)
    lval.append( row[2] )
    if row[0] == 2014:
        print row
    
cursor.execute("""
  WITH ranks as (
  SELECT extract(year from day + '4 months'::interval) as yr, extract(doy from day) as doy, high, 
  rank() OVER (PARTITION by extract(year from day + '4 months'::interval) ORDER by high ASC) from alldata_ia where station = 'IA2203'
  and year > 1892
  ORDER by yr ASC
  )

  SELECT yr, doy, high from ranks WHERE rank = 1

""")
for row in cursor:
    addon = 366 if row[1] < 180 else 0
    hdoy.append( row[1] + addon )
    hval.append( row[2] )
    if row[0] == 2014:
        print row
```

```python
import datetime
sts = datetime.datetime(1999,1,1)
ets = datetime.datetime(2001,1,1)
interval = datetime.timedelta(days=1)
xticks = []
xticklabels = []
i = 1
while sts < ets:
    if sts.day == 1:
        xticks.append(i)
        xticklabels.append( sts.strftime("%b") )
    i += 1
    sts += interval
import matplotlib.pyplot as plt

(fig, ax) = plt.subplots(1,1)

ax.scatter(hdoy, hval, marker='+', color='g', s=50, label='Min High')
ax.scatter(ldoy, lval, marker='o', color='b', s=50, label='Min Low')
ax.scatter(ldoy[-1], lval[-1], marker='o', s=50, color='r', edgecolor='r', zorder=2)
ax.scatter(hdoy[-1], hval[-1], marker='+', s=50, color='r', zorder=2)
ax.set_xticks(xticks)
ax.legend(loc=3)
ax.set_xticklabels(xticklabels)
ax.set_ylabel("Temperature $^\circ$F")
ax.set_title("1893-2013 Des Moines\nDate and Value of Minimum Hi/Lo Temperature")
ax.set_xlabel("red symbols are 2013-2014 winter thru 13 Jan")
ax.grid()
ax.set_xlim(300,460)
```
