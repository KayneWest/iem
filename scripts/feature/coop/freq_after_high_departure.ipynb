```python
import psycopg2
DBCONN = psycopg2.connect(database='coop', host='mesonet.agron.iastate.edu', user='nobody')
cursor = DBCONN.cursor()

cursor.execute("""with obs as 
 (select month, sday, high, snow, max(snow) over 
 (ORDER by day ASC rows between 1 following and 3 following)
 from alldata_ia where station = 'IA2203' and day > '1893-01-01'), 
 climo as (select to_char(valid,'mmdd') as sday, high from climate51 
 where station = 'IA2203') 
 
 SELECT round((obs.high - climo.high)::numeric,0) as diff, 
 sum(case when max >= 0.1 then 1 else 0 end), 
 count(*) from obs JOIn climo on (climo.sday = obs.sday) 
 WHERE obs.month in (12,1,2) GROUP by diff ORDER by diff ASC""")

ratios = []
diff = []
hits = 0
total = 0
for row in cursor:
    hits += row[1]
    total += row[2]
    diff.append( float(row[0]) )
    ratios.append( float(row[1] / float(row[2]) * 100.0) )
```

```python
import matplotlib.pyplot as plt

(fig, ax) = plt.subplots(1,1)

ax.bar(diff, ratios, fc='b', ec='b')
ax.set_xlabel("High Temperature Departure $^\circ$F")
ax.grid(True)
ax.set_xlim(-45,45)
ax.set_ylabel("Snowfall Observed [%]")
ax.set_title("Frequency of Snowfall Observation [DJF] over Three Days\nafter day of high temp departure (Des Moines 1893-2012)") 
ax.axhline( hits / float(total) * 100, lw=2, color='k')
ax.text( 10, hits / float(total) * 100 + 2, "%.0f%% Overall Avg" % (hits / float(total) * 100,))
```
