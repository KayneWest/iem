```python
txt = """WAUKON                       12.0 IN   0755 AM 03/23   43.27N/91.48W        

...CHICKASAW COUNTY...
N. WASHINGTON                5.9 IN    0707 AM 03/23   43.12N/92.42W        
NEW HAMPTON                  4.5 IN    0700 AM 03/23   43.06N/92.31W        
NEW HAMPTON 0.4 SW           4.0 IN    0700 AM 03/23   43.06N/92.32W        
IONIA 2W                     2.0 IN    0700 AM 03/23   43.03N/92.50W        

...CLAYTON COUNTY...
MARQUETTE                    12.0 IN   0730 AM 03/23   43.04N/91.18W        
MONONA WWTP                  10.8 IN   0700 AM 03/23   43.04N/91.40W        
VOLGA 1NE                    3.3 IN    0700 AM 03/23   42.81N/91.52W        
ELKADER 6SSW                 1.8 IN    0700 AM 03/23   42.78N/91.45W        

...FAYETTE COUNTY...
WAUCOMA 3.2 S                3.8 IN    0700 AM 03/23   43.01N/92.03W        
FAYETTE                      2.4 IN    0700 AM 03/23   42.85N/91.82W        
WADENA                       1.5 IN    0800 AM 03/23   42.84N/91.66W        

...FLOYD COUNTY...
CHARLES CITY                 1.5 IN    0700 AM 03/23   43.06N/92.67W        
NORA SPRINGS 2.4 SSE         1.5 IN    0700 AM 03/23   43.11N/92.99W        
NASHUA 2SW                   0.7 IN    0700 AM 03/23   42.94N/92.57W        

...HOWARD COUNTY...
CRESCO 1NE                   5.8 IN    0700 AM 03/23   43.39N/92.09W        
ELMA                         2.9 IN    0700 AM 03/23   43.24N/92.44W        

...MITCHELL COUNTY...
RICEVILLE                    4.0 IN    0700 AM 03/23   43.36N/92.55W        
ST ANSGAR                    2.8 IN    0630 AM 03/23   43.38N/92.92W        
OSAGE                        2.6 IN    0700 AM 03/23   43.28N/92.81W        
2 S MONA                     2.5 IN    1015 AM 03/23   43.46N/92.94W        

...WINNESHIEK COUNTY...
8 ENE DECORAH                11.5 IN   0730 AM 03/23   43.35N/91.65W        
7 NE DECORAH                 11.5 IN   0730 AM 03/23   43.38N/91.69W        
DECORAH 7.9 ENE              11.5 IN   0728 AM 03/23   43.34N/91.64W        
DECORAH 5E                   9.5 IN    0730 AM 03/23   43.30N/91.70W        

...MINNESOTA...

...DODGE COUNTY...
MANTORVILLE 2ESE             8.0 IN    1000 AM 03/23   44.05N/92.72W        
KASSON 1NE                   8.0 IN    0600 AM 03/23   44.05N/92.74W        

...FILLMORE COUNTY...
SPRING VALLEY 3E             10.0 IN   0700 AM 03/23   43.68N/92.33W        
SPRING VALLEY 3SW            10.0 IN   0830 AM 03/23   43.66N/92.43W        
PRESTON 1SSE                 9.6 IN    0830 AM 03/23   43.66N/92.08W        
HARMONY                      9.0 IN    0800 AM 03/23   43.55N/92.01W        
CHATFIELD 9.0 ESE            8.2 IN    0700 AM 03/23   43.79N/92.03W        
3 N FILLMORE                 8.0 IN    0800 AM 03/23   43.79N/92.26W        
MABEL                        8.0 IN    0700 AM 03/23   43.52N/91.76W        
LANESBORO                    7.9 IN    0700 AM 03/23   43.72N/91.97W        
PETERSON 1S                  7.7 IN    0704 AM 03/23   43.78N/91.82W        
WYKOFF 3.1 NW                7.6 IN    0700 AM 03/23   43.74N/92.31W        
2 SSW WHALAN                 7.4 IN    1000 AM 03/23   43.71N/91.94W        
PRESTON                      7.4 IN    0815 AM 03/23   43.67N/92.07W        
WYKOFF                       6.5 IN    0700 AM 03/23   43.71N/92.27W        
LANESBORO                    4.0 IN    0700 AM 03/23   43.72N/91.97W        

...HOUSTON COUNTY...
CALEDONIA 5.4 S              8.0 IN    0800 AM 03/23   43.56N/91.51W        
1 NNW LA CRESCENT            7.0 IN    0700 AM 03/23   43.83N/91.31W        
CALEDONIA                    6.7 IN    0700 AM 03/23   43.63N/91.50W        
SPRING GROVE 4N              6.0 IN    0700 AM 03/23   43.61N/91.62W        
LA CRESCENT 0.7 NNW          4.6 IN    0730 AM 03/23   43.84N/91.31W        

...MOWER COUNTY...
GRAND MEADOW                 8.5 IN    0700 AM 03/23   43.71N/92.57W        
AUSTIN                       6.0 IN    0800 AM 03/23   43.67N/92.97W        
AUSTIN WASTE WATER TREATMENT 5.7 IN    0700 AM 03/23   43.65N/92.97W        
AUSTIN 2NE                   5.2 IN    0930 AM 03/23   43.68N/92.95W        

...OLMSTED COUNTY...
PINE ISLAND 2S               10.6 IN   0730 AM 03/23   44.19N/92.65W        
ORONOCO                      10.5 IN   0630 AM 03/23   44.17N/92.53W        
ROCHESTER 2.6 NE             10.0 IN   0600 AM 03/23   44.04N/92.44W        
ORONOCO 3SE                  9.9 IN    0915 AM 03/23   44.14N/92.48W        
ORONOCO 1S                   9.8 IN    0700 AM 03/23   44.15N/92.54W        
BYRON 4 NORTH                8.4 IN    0700 AM 03/23   44.09N/92.64W        
ROCHESTER                    8.0 IN    0735 AM 03/23   44.01N/92.48W        
ROCHESTER AP 2NE             7.9 IN    0700 AM 03/23   43.93N/92.48W        
ROCHESTER INTERNATIONAL AIRP 7.9 IN    0654 AM 03/23   43.91N/92.50W        
ELGIN 2SSW                   7.6 IN    0700 AM 03/23   44.10N/92.27W        

...WABASHA COUNTY...
5 NNE ZUMBRO FALLS           10.5 IN   0830 AM 03/23   44.35N/92.39W        
1 S HAMMOND                  10.5 IN   0645 AM 03/23   44.21N/92.37W        
ZUMBRO FALLS 3.5 SSW         9.3 IN    0745 AM 03/23   44.24N/92.46W        
PLAINVIEW                    9.0 IN    0730 AM 03/23   44.16N/92.17W        
THEILMAN 1SSW                8.9 IN    0830 AM 03/23   44.28N/92.19W        
LAKE CITY                    8.0 IN    0800 AM 03/23   44.44N/92.28W        
WABASHA                      7.8 IN    0700 AM 03/23   44.39N/92.05W        

...WINONA COUNTY...
WINONA 4SE                   9.5 IN    0730 AM 03/23   44.02N/91.61W        
GOODVIEW                     8.9 IN    0600 AM 03/23   44.07N/91.71W        
MINNESOTA CITY               8.0 IN    0815 AM 03/23   44.09N/91.75W        
WINONA 3S                    7.5 IN    0800 AM 03/23   43.99N/91.61W        
ALTURA 5W                    7.0 IN    0700 AM 03/23   44.06N/92.04W        
MINNESOTA CITY DAM 5         6.0 IN    0700 AM 03/23   44.16N/91.81W        
TREMPEALEAU DAM 6            6.0 IN    0600 AM 03/23   44.00N/91.44W        
WINONA DAM 5A                6.0 IN    0600 AM 03/23   44.09N/91.67W        
LA CRESCENT DAM 7            4.0 IN    0600 AM 03/23   43.87N/91.31W        

...WISCONSIN...

...ADAMS COUNTY...
FRIENDSHIP                   0.3 IN    0700 AM 03/23   43.98N/89.83W        

...BUFFALO COUNTY...
ALMA DAM 4                   7.2 IN    0600 AM 03/23   44.33N/91.92W        

...CLARK COUNTY...
NEILLSVILLE 3ESE             0.7 IN    0700 AM 03/23   44.54N/90.54W        
NEILLSVILLE 3NW              0.1 IN    0729 AM 03/23   44.61N/90.65W        

...CRAWFORD COUNTY...
PRAIRIE DU CHIEN             13.0 IN   0700 AM 03/23   43.05N/91.13W        
STEUBEN 4SE                  10.0 IN   0700 AM 03/23   43.13N/90.84W        
LYNXVILLE DAM 9              10.0 IN   0600 AM 03/23   43.21N/91.10W        
MOUNT ZION                   8.8 IN    0730 AM 03/23   43.26N/90.73W        
GAYS MILLS 6.1 SE            8.8 IN    0700 AM 03/23   43.26N/90.76W        

...GRANT COUNTY...
WOODMAN                      15.0 IN   1030 AM 03/23   43.09N/90.80W        
FENNIMORE                    15.0 IN   0734 AM 03/23   42.98N/90.66W        
LANCASTER                    11.0 IN   0700 AM 03/23   42.85N/90.71W        
LANCASTER 0.3 N              11.0 IN   0700 AM 03/23   42.85N/90.71W        
MUSCODA                      9.5 IN    0800 AM 03/23   43.20N/90.44W        
LANCASTER 4WSW               8.0 IN    0800 AM 03/23   42.83N/90.79W        
CUBA CITY                    3.0 IN    0800 AM 03/23   42.60N/90.43W        
SINSINAWA                    1.5 IN    1037 AM 03/23   42.52N/90.54W        
CASSVILLE                    1.5 IN    0800 AM 03/23   42.72N/90.99W        

...JACKSON COUNTY...
MELROSE                      4.3 IN    0529 AM 03/23   44.13N/90.99W        

...LA CROSSE COUNTY...
HOLMEN 1NW                   8.5 IN    0600 AM 03/23   43.97N/91.27W        
WEST SALEM                   7.0 IN    0700 AM 03/23   43.90N/91.09W        
LA CROSSE REGIONAL AIRPORT   7.0 IN    0653 AM 03/23   43.88N/91.26W        
WFO LA CROSSE                6.7 IN    0600 AM 03/23   43.82N/91.19W        
HOLMEN 2S                    6.5 IN    0700 AM 03/23   43.93N/91.25W        
LA CROSSE 4NNW               6.4 IN    0730 AM 03/23   43.87N/91.27W        
ONALASKA                     6.1 IN    0800 AM 03/23   43.91N/91.23W        
ONALASKA 1.6 NW              6.0 IN    0700 AM 03/23   43.90N/91.24W        
BARRE MILLS 2N               5.5 IN    0700 AM 03/23   43.85N/91.11W        
BANGOR                       5.3 IN    0800 AM 03/23   43.89N/90.99W        
LA CROSSE 1SE                5.0 IN    0600 AM 03/23   43.81N/91.22W        
FOUR CORNERS                 4.0 IN    0700 AM 03/23   44.07N/90.92W        
LA CROSSE SNOW               2.2 IN    0600 AM 03/23   43.86N/91.27W        

...MONROE COUNTY...
SPARTA                       4.6 IN    0800 AM 03/23   43.94N/90.82W        
SPARTA HIGHWAY DEPT          3.0 IN    0800 AM 03/23   43.94N/90.80W        
TUNNEL CITY                  2.9 IN    0700 AM 03/23   44.01N/90.57W        

...RICHLAND COUNTY...
BOAZ 2NW                     7.5 IN    0800 AM 03/23   43.36N/90.56W        
RICHLAND CENTER              6.5 IN    0558 AM 03/23   43.34N/90.38W        
RICHLAND CENTER 4N           6.0 IN    0700 AM 03/23   43.40N/90.39W        
ITHACA 2S                    5.5 IN    0615 AM 03/23   43.34N/90.28W        

...TREMPEALEAU COUNTY...
ETTRICK 4WNW                 6.6 IN    0637 AM 03/23   44.19N/91.34W        
1 N GALESVILLE               6.0 IN    0730 AM 03/23   44.10N/91.36W        
ARCADIA                      5.5 IN    0649 AM 03/23   44.25N/91.49W        
WHITEHALL 2W                 4.0 IN    1000 AM 03/23   44.37N/91.36W        

...VERNON COUNTY...
3 NNW READSTOWN              9.8 IN    0630 AM 03/23   43.49N/90.79W        
READSTOWN 4NE                6.8 IN    0700 AM 03/23   43.48N/90.69W        
STODDARD                     6.5 IN    0700 AM 03/23   43.66N/91.22W        
GENOA DAM 8                  6.5 IN    0600 AM 03/23   43.57N/91.23W        
VIROQUA                      6.0 IN    0700 AM 03/23   43.56N/90.88W        
VIROQUA 0.8 ESE              6.0 IN    0700 AM 03/23   43.55N/90.87W        
WESTBY 3ENE                  5.3 IN    0710 AM 03/23   43.67N/90.81W        
LA FARGE                     4.5 IN    0700 AM 03/23   43.57N/90.64W        
STODDARD 5NNE                4.5 IN    0700 AM 03/23   43.72N/91.17W        
WESTBY 3ENE                  4.2 IN    0700 AM 03/23   43.67N/90.81W        
VALLEY                       4.0 IN    0830 AM 03/23   43.64N/90.54W        
2 NE VALLEY                  4.0 IN    0700 AM 03/23   43.66N/90.51W        
HILLSBORO WSW                3.0 IN    0700 AM 03/23   43.65N/90.35W  """

def get_lsrs(sts, ets):
    ''' Return snowfall lsrs '''
    POSTGIS = psycopg2.connect(database='postgis', host='localhost', port=5555, user='nobody')
    cursor = POSTGIS.cursor()
    data = []
    cursor.execute("""
    SELECT state, max(magnitude) as val, ST_x(geom) as lon, ST_y(geom) as lat
    from lsrs WHERE type in ('S') and magnitude > 0 and 
    valid >= %s and valid < %s and wfo in ('FSD', 'ARX', 'DVN', 'EAX', 'OAX', 'DMX')
    GROUP by state, lon, lat
    """, (sts, ets))
    for i, row in enumerate(cursor):
        data.append(dict(val=row[1], lat=row[3], lon=row[2], label='l%s' % (i,), source='lsr', useme=True))
    
    cursor.close()
    POSTGIS.close()
    return data

def get_coop(dates):
    ''' Return snowfall obs for the given dates '''
    if len(dates) == 1:
        dates.append( dates[0] )
    IEM = psycopg2.connect(database='iem', host='localhost', port=5555, user='nobody')
    cursor = IEM.cursor()
    data = []
    cursor.execute("""
    WITH obs as (
        SELECT s.iemid, count(*), sum(snow) from summary s JOIN stations t on (t.iemid = s.iemid)
        WHERE t.network in ('IACOCORAHS','IA_COOP') and s.day in """+str(tuple(dates))+""" 
        and snow >= 0 GROUP by s.iemid
    )
    SELECT id, obs.count, obs.sum, ST_x(geom), ST_y(geom), county from stations JOIN obs
    on (obs.iemid = stations.iemid)
    """)
    for row in cursor:
        #if row[1] == 2:
        val = row[2]
        #if row[3] > -93.6 and row[4] > 42:
        #    val = 0
        data.append(dict(val=val, lat=row[4], lon=row[3], label=row[0], source='coop', useme=True))
    
    cursor.close()
    IEM.close()
    return data

crows = get_coop( ['2015-03-23'] )
sts = datetime.datetime(2015,3,23, 0)
sts = sts.replace(tzinfo=pytz.timezone("UTC"))
ets = datetime.datetime(2015,3, 27,0,58)
ets = ets.replace(tzinfo=pytz.timezone("UTC"))
lrows = get_lsrs(sts, ets)

prows = []
for line in txt.split("\n"):
    if line.strip() == "" or line[0] == ".":
        continue
    val = float(line[29:33])
    label = line[:25].strip()
    lon = 0 - float(line[62:67])
    lat = float(line[55:60])
    prows.append(dict(val=val,lat=lat,lon=lon,label=label,source='pns',useme=True))
    
df = pd.DataFrame(crows + lrows + prows)
```

```python
grid = np.ones( (iemre.NX, iemre.NY), 'f') * -1.0
stations = {}

gx = np.digitize(df['lon'], iemre.XAXIS) - 1
gy = np.digitize(df['lat'], iemre.YAXIS) - 1

for i, row in df.iterrows():
    if gx[i] == -1:
        gx[i] = 0
    if gy[i] == -1:
        gy[i] = 0
    if float(row['val']) > grid[gx[i], gy[i]]:
        grid[gx[i], gy[i]] = row['val']
        stations["%s_%s" % (gx[i], gy[i])] = row['label']

#print grid
newrows = []
for i in range(iemre.NX):
    for j in range(iemre.NY):
        if grid[i,j] >= 0:
            newrows.append(dict(lon=iemre.XAXIS[i], lat=iemre.YAXIS[j], val=float(grid[i,j]), useme=True,
                                label=stations["%s_%s" % (i,j)]))
            #print i, j, float(grid[i,j])
            
df2 = pd.DataFrame(newrows)
        
```

```python
for i, row in df2.iterrows():
    if row['label'] in ('TRPI4', 'MHRI4', 'l148', 'l164') or row['val'] > 16.5:
        df2['useme'][i] = False
    else:
        df2['useme'][i] = True
cmap = whitebluegreenyellowred()
m = MapPlot(figsize=(20,18))
df3 = df2[df2.useme]
m.contourf(np.array(df3.lon), np.array(df3.lat), np.array(df3.val), np.arange(0,15,0.75), cmap=cmap)
m.plot_values(df3['lon'], df3['lat'], df3['val'], '%.2f', labels=df3.label)
```
