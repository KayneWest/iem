```python
import psycopg2
DBCONN = psycopg2.connect(database='coop', host='mesonet.agron.iastate.edu', user='nobody')
cursor = DBCONN.cursor()

cursor.execute("""SELECT day, high from alldata_ia where station = 'IA2203' ORDER by day ASC""")
days = []
highs = []
for row in cursor:
    days.append( row[0] )
    highs.append( row[1] )
```

```python
import numpy as np
sums = np.zeros( (52), 'f')
cnts = np.zeros( (52), 'f')
sz = len(days)
for i, (d, h) in enumerate(zip(days, highs)):
    if h < -20 or h >= 32:
        continue
    idx = h + 20
    j = i+1
    while j < sz and highs[j] < 33:
        j += 1
    sums[idx] += (j-i)-1
    cnts[idx] += 1
    
```

```python
import matplotlib.pyplot as plt

(fig, ax) = plt.subplots(1,1)

ax.bar( np.arange(-20,32)-0.5, sums / cnts )
ax.set_xlim(-10.5,32)
ax.grid(True)
ax.set_xlabel("Daily High Temperature $^\circ$F")
ax.set_ylabel("Number of Days")
ax.set_title("Mean number of days to high above 32$^\circ$F\nafter day with given high 1886-2012 Des Moines")
```
