```python
import psycopg2
DBCONN = psycopg2.connect(database='coop', host='mesonet.agron.iastate.edu', user='nobody')
cursor = DBCONN.cursor()

cursor.execute("""SELECT day, high from alldata_ia where station = 'IA2203' ORDER by day ASC""")
days = []
highs = []
for row in cursor:
    days.append( row[0] )
    highs.append( row[1] )
```

```python
import numpy as np
sums = np.zeros( (52), 'f')
cnts = np.zeros( (52), 'f')
sz = len(days)
for i, (d, h) in enumerate(zip(days, highs)):
    if h < -20 or h >= 32:
        continue
    idx = h + 20
    j = i+1
    while j < sz and highs[j] < 50:
        j += 1
    sums[idx] += (j-i)-1
    cnts[idx] += 1
    
```

```python
import matplotlib.pyplot as plt
from scipy import stats

h_slope, intercept, h_r_value, p_value, std_err = stats.linregress(np.arange(0,32), sums[20:]/cnts[20:])

(fig, ax) = plt.subplots(1,1)

ax.bar( np.arange(-20,32)-0.5, sums / cnts , ec='b', fc='b')
print h_slope, intercept
ax.plot([-10,32], [h_slope * -10.0 + intercept, h_slope * 32.0 + intercept], color='k', lw=2, zorder=2)
ax.set_xlim(-10.5,32)
ax.set_ylim(0,45)
ax.grid(True)
ax.set_xlabel("Daily High Temperature $^\circ$F")
ax.set_ylabel("Number of Days")
ax.set_title("Mean number of days to high at or above 50$^\circ$F\nafter day with given high 1886-2012 Des Moines")
```
