```python
import psycopg2
import numpy as np
import datetime

COOP = psycopg2.connect(database='coop', host='mesonet.agron.iastate.edu', user='nobody')
cursor = COOP.cursor()

cursor.execute("""
 select sday, max(high - max) from (select sday, day, high, 
 max(high) over (ORDER by day ASC rows between 13 preceding and 1 preceding) from 
 alldata_ia where station = 'IA2203' ORDER by day ASC) as foo GROUP by sday ORDER by sday ASC
""")
days = []
vals = []
for row in cursor:
    ts = datetime.datetime.strptime("2000%s" % (row[0],), '%Y%m%d')
    days.append(ts.date())
    vals.append( row[1] )
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pytz

(fig, ax) = plt.subplots(1,1)

ax.bar(days, vals, fc='r', ec='r')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%-d\n%b'))
ax.grid(True)
ax.set_title("Des Moines Maximum Daily High Temp Change\npositive change from maximum over previous 13 days")
ax.set_ylabel("High Temp Change $^\circ$F")
```
