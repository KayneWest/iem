```python
import psycopg2
from pyiem.plot import MapPlot
import matplotlib.cm as cm
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
COOP = psycopg2.connect(database='coop', host='iemdb', user='nobody')
cursor = COOP.cursor()

cursor.execute("""
WITH obs as (select station, avg((high+low)/2.) from alldata WHERE day >= '2013-12-01'
 GROUP by station),
 
climo as (SELECT station, avg((high+low)/2.) from climate51 where valid >= '2000-12-01' 
or valid < '2000-02-12'
 GROUP by station)
 
SELECT o.station, o.avg - c.avg, ST_x(t.geom), ST_y(t.geom) from obs o, climo c, stations t
 WHERE t.id = o.station and o.station = c.station 
""")

vals = []
for row in cursor:
    if row[0][2] == 'C' or row[0][2:] == '0000':
        continue
    vals.append( dict(stid=row[0], lat=row[3], lon=row[2], val=row[1], useme=True) )

df = pd.DataFrame(vals)
```

```python
for i, row in df.iterrows():
    if row['val'] > -1:
        df['useme'][i] = False
    else:
        df['useme'][i] = True
df2 = df[df.useme]
m = MapPlot(sector='midwest', title='1 Dec 2013 - 11 Feb 2014 Average Temperature Departure $^\circ$F')
cmap = cm.get_cmap('RdYlBu_r')

#m.plot_values(df2.lon, df2.lat, df2.val, '%.0f', textsize=8)
CS = m.contourf(np.array(df2.lon), np.array(df2.lat), np.array(df2.val), np.arange(-12,13,2), cmap=cmap,
           units='$^\circ$F')
#plt.clabel(CS, fmt = '%2.1f', colors = 'w', fontsize=14)
```
