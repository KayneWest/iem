```python
%pylab inline
import psycopg2
COOP = psycopg2.connect(database='coop', host='iemdb', user='nobody')
cursor = COOP.cursor()

dates = []
lows = []
cursor.execute("""SELECT day, low from alldata_ia WHERE station = 'IA2203'
  ORDER by day ASC""")
for row in cursor:
    dates.append( row[0] )
    lows.append( row[1] )

COOP.close()
```

```python
years = []
steps = []
steps32 = []
minlow = 99
step = 0
step2 = 0
for d,l in zip(dates, lows):
    if d.month == 1 and d.day == 1:
        years.append( d.year - 1)
        steps.append( step )
        steps32.append( step2 )
    if d.month < 7:
        continue
    if d.month == 7 and d.day == 1:
        minlow = 50
        step = 0
        step2 = 0
    if l < minlow:
        if d.year == 2013:
            print minlow, l, d, minlow, step, step2
        minlow = l
        step += 1
        if l > 32:
            step2 += 1
            
years.append( d.year )
steps.append( step )
steps32.append( step2 )
```

```python
import matplotlib.pyplot as plt

(fig, ax) = plt.subplots(1,1)
ax.bar(years[1:-2], steps[1:-2], fc='r', ec='r', zorder=1, label='All Events')
ax.bar(years[1:-2], steps32[1:-2], fc='b', ec='b', zorder=2, label='to 32$^\circ$F')
ax.grid(True)
ax.legend(loc=2,ncol=2, fontsize=10)
ax.set_xlim(1879,2013)
ax.set_title("1880-2012 Des Moines Days (low below 50$^\circ$F)\n with New Coldest Temp of Jul-Dec Season") 

fig.savefig('test.ps')
import iemplot
iemplot.makefeature('test')
```
