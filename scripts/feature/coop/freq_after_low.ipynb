```python
import psycopg2
DBCONN = psycopg2.connect(database='coop', host='mesonet.agron.iastate.edu', user='nobody')
cursor = DBCONN.cursor()

cursor.execute("""with obs as 
 (select month, sday, low, max(high) over 
 (ORDER by day ASC rows between 1 following and 3 following)
 from alldata_ia where station = 'IA2203' and day > '1893-01-01')
 
 SELECT low, 
 sum(case when max >= 32 then 1 else 0 end), 
 count(*) from obs  
 WHERE obs.month in (12,1,2) GROUP by low ORDER by low ASC""")

ratios = []
lows = []
hits = 0
total = 0
for row in cursor:
    hits += row[1]
    total += row[2]
    lows.append( float(row[0]) )
    ratios.append( float(row[1] / float(row[2]) * 100.0) )
```

```python
import matplotlib.pyplot as plt

(fig, ax) = plt.subplots(1,1)

ax.bar(lows, ratios, fc='b', ec='b')
ax.set_xlabel("Low Temperature $^\circ$F")
ax.grid(True)
ax.set_xlim(-30,35)
ax.set_ylabel("At or Above Freezing Temp Observed [%]")
ax.set_title("Frequency of Above Freezing Temp [DJF] over Three Days\nafter day with low temperature(Des Moines 1893-2012)") 
ax.axhline( hits / float(total) * 100, lw=2, color='k')
ax.text( -20, hits / float(total) * 100 + 2, "%.0f%% Overall Avg" % (hits / float(total) * 100,))
```
