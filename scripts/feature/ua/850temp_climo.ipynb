```python
import psycopg2
pgconn = psycopg2.connect(database='postgis', host='iemdb', user='nobody')
cursor = pgconn.cursor()
cursor.execute("""SET TIME ZONE 'UTC'""")
cursor.execute("""
 SELECT w, stddev(data), avg(data), min(data), max(data), 
 max(case when ntile = 5 then data else -999 end),
 max(case when ntile = 95 then data else -999 end) from
 
 (select extract(week from valid) as w, tmpc as data, 
 ntile(100) OVER (PARTITION by extract(week from valid) order by tmpc ASC) from 
 raob_profile p JOIN raob_flights f on (f.fid = p.fid) WHERE extract(hour from valid) = 12 and 
 f.station in ('KOAX', 'KOVN', 'KOMA') and p.pressure = 850 and tmpc is not null) as foo
 
 GROUP by w ORDER by w ASC
""")

avg = []
stddev = []
week = []
maxes = []
mines = []
n5 = []
n95 = []
for row in cursor:
    week.append( row[0] )
    stddev.append( row[1] )
    avg.append( row[2] )
    mines.append( row[3] )
    maxes.append( row[4] )
    n5.append( row[5] )
    n95.append( row[6] )
```

```python
import numpy as np
avg = np.array(avg)
week = np.array(week) - 1
stddev = np.array(stddev)
mines = np.array(mines)
maxes = np.array(maxes)
n5 = np.array(n5)
n95 = np.array(n95)
```

```python
cursor.execute("""
 select extract(doy from valid), tmpc from 
 raob_profile_2015 p JOIN raob_flights f on (f.fid = p.fid) WHERE extract(hour from valid) = 12 and 
 f.station in ('KOAX', 'KOVN', 'KOMA') and p.pressure = 850 and tmpc is not null 
 ORDER by valid ASC
""")
doy = []
tmpc = []
for row in cursor:
    doy.append( row[0] )
    tmpc.append( row[1] )
```

```python
def fill_between(x, y1, y2=0, ax=None, **kwargs):
    """Plot filled region between `y1` and `y2`.

    This function works exactly the same as matplotlib's fill_between, except
    that it also plots a proxy artist (specifically, a rectangle of 0 size)
    so that it can be added it appears on a legend.
    """
    ax = ax if ax is not None else plt.gca()
    ax.fill_between(x, y1, y2, **kwargs)
    p = plt.Rectangle((0, 0), 0, 0, **kwargs)
    ax.add_patch(p)
    return p

import matplotlib.pyplot as plt

(fig, ax) = plt.subplots(1,1)

fill_between(week * 7, mines, maxes, color='#EEEEEE', zorder=1, label='Range')
#ax.fill_between(week * 7, avg - (stddev *2.), avg + (stddev * 2.), color='lightblue', zorder=2)
fill_between(week * 7, avg - (stddev *1.), avg + (stddev * 1.), color='tan', zorder=3,
             label="+/- $\sigma$")
ax.plot(week * 7, avg, lw=2, color='k', zorder=4, label='Avg')
ax.plot(week * 7, mines, lw=2, color='brown', zorder=4)
ax.plot(week * 7, maxes, lw=2, color='brown', zorder=4)
ax.plot(week * 7, n5, lw=2, color='g', zorder=4, label="5 & 95 p'tile")
ax.plot(week * 7, n95, lw=2, color='g', zorder=4)
ax.plot(doy, tmpc, zorder=5, label='2015')
ax.set_ylim(-35,50)
ax.set_xticks( (1,32,60,91,121,152,182,213,244,274,305,335,365) )
ax.set_xticklabels( ('Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec') )
ax.set_xlim(0,(max(week)-2.) * 7)
ax.set_title("1960-2014 12 UTC Omaha\n850 hPa Temperature Climatology")
ax.set_ylabel("Temperature $\circ$C")

ax.legend(ncol=3, fontsize=10)
ax.grid(True)
```
