```python
import datetime
import psycopg2
import numpy as np
RWIS = psycopg2.connect(database='rwis', host='iemdb', user='nobody')
cursor = RWIS.cursor()

data = np.zeros( (15, 31+31+13), 'f')

cursor.execute("""
 select date(valid) as d, avg(s0temp), avg(s1temp), avg(s2temp), avg(s3temp),
 avg(s4temp), avg(s5temp), avg(s6temp), avg(s7temp),
 avg(s8temp), avg(s9temp), avg(s10temp), avg(s11temp),
 avg(s12temp), avg(s13temp), avg(s14temp)
 from alldata_soil where station = 'RTPI4' and valid > '2013-12-01' 
 and valid < '2014-02-14' GROUP by d ORDER by d ASC
""")
sts = datetime.date(2013,12,1)
xticks = []
xticklabels = []
for i, row in enumerate(cursor):
    if row[0].day == 1:
        xticks.append(i)
        xticklabels.append( row[0].strftime("%-d %b\n%Y") )
    day = int((row[0] - sts).days)
    data[:,day] = row[1:]
```

```python
import matplotlib.pyplot as plt
import matplotlib.mpl as mpl

depths = np.array([1,	3,	6,	9,	12,	18,	24,	30,	36,	42,	48,	54,	60,	66,	72])
days = np.arange(31+31+13)

bounds = np.arange(8,49,4)
cmap = mpl.cm.jet
norm = mpl.colors.BoundaryNorm(bounds, cmap.N)

(fig, ax) = plt.subplots(1,1)

C = ax.pcolormesh(days, depths, data, cmap=cmap, norm=norm)
ax.set_xticks(xticks)
ax.set_xticklabels(xticklabels)
ax.grid(True)
ax.set_ylabel("Depth [inches]")
ax.set_title("Tipton DOT RWIS Subsurface Temperature $^\circ$F Probe\n1 Dec 2013 - 13 Feb 2014 Below Pavement in Roadbed")
fig.colorbar(C)
ax.set_ylim(74,0)
ax.set_xlim(-0.5, 31+31+12)
```
