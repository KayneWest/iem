```python
import psycopg2
from shapely.wkt import loads

POSTGIS = psycopg2.connect(database='postgis', host='iemdb', user='nobody')
cursor = POSTGIS.cursor()

roads = []
freq = []

cursor.execute("""
WITH obs as (
 SELECT segid, sum(valid - lag) from
 (SELECT segid, valid,  lag(valid) OVER (PARTITION by segid ORDER by valid ASC),
 cond_code from roads_2014_log WHERE valid > '2014-01-01') as foo
 WHERE cond_code > 0 GROUP by segid
)
 SELECT obs.sum, ST_AsEWKT(ST_Transform(base.simple_geom,4326)) from obs JOIN
 roads_base base on (obs.segid = base.segid) 

""")

for row in cursor:
    if row[1] is None:
        continue
    roads.append(loads(row[1][10:]))
    freq.append(row[0])
```

```python
import datetime

sts = datetime.datetime(2014,1,1)
ets = datetime.datetime(2014,2,6,18)
period = (ets - sts).days * 86400. + (ets - sts).seconds
```

```python
from pyiem.plot import MapPlot, whitebluegreenyellowred
import numpy as np
import matplotlib.colors as mpcolors

cmap = whitebluegreenyellowred()
clevs = np.arange(0,101,10)
norm = mpcolors.BoundaryNorm(clevs, cmap.N)

m = MapPlot(title='1 Jan - 6 Feb 2014 Non-Normal Driving Condition Percentage',
            subtitle='Based on Iowa DOT Road Condition Reports')

for geoms, f in zip(roads,freq):
    dur = f.days * 86400. + f.seconds
    percent = (dur / period) * 100.
    for geo in geoms:
        x,y = geo.xy
        x,y = m.map(x, y)
        m.ax.plot(x,y, lw=3., color=cmap( norm([percent,]) )[0])

m.draw_colorbar(clevs, cmap, norm, units='percent')
        
m.postprocess(filename='test.ps')
import iemplot
iemplot.makefeature('test')
```
